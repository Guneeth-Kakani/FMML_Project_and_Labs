# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
def plot_KNN_error_rate(xdata,ydata):
    error_rate = []
    test_scores = []
    train_scores = []

    ## [REQUIRED] Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=101)

    for i in range(1,40):
        ## [REQUIRED] Complete the code in the next three lines
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train, y_train)
        pred_i = knn.predict(X_test)

        error_rate.append(np.mean(pred_i != y_test))
        train_scores.append(knn.score(X_train,y_train))
        test_scores.append(knn.score(X_test,y_test))

    plt.figure(figsize=(12,8))
    plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
            markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show() # Added plt.show() for visual output
    print()
    ## score that comes from testing on the same datapoints that were used for training
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
    print()
    ## score that comes from testing on the datapoints that were split in the beginning to be used for testing solely
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))

    return test_scores
```

```python
#@title Answer to Task-6. Refer to and run this only if you are unable to complete the task in the previous cell.
def plot_KNN_error_rate(xdata,ydata):
  error_rate = []
  test_scores = []
  train_scores = []

  X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=101)

  for i in range(1,40):
      knn = KNeighborsClassifier(n_neighbors=i)
      knn.fit(X_train, y_train)
      pred_i = knn.predict(X_test)

      error_rate.append(np.mean(pred_i != y_test))
      train_scores.append(knn.score(X_train,y_train))
      test_scores.append(knn.score(X_test,y_test))

  plt.figure(figsize=(12,8))
  plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
          markerfacecolor='red', markersize=10)
  plt.title('Error Rate vs. K Value')
  plt.xlabel('K')
  plt.ylabel('Error Rate')
  print()
  ## score that comes from testing on the same datapoints that were used for training
  max_train_score = max(train_scores)
  train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
  print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
  print()
  ## score that comes from testing on the datapoints that were split in the beginning to be used for testing solely
  max_test_score = max(test_scores)
  test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
  print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))

  return test_scores
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# NOTE: The 'diabetes_data' DataFrame is assumed to be loaded and 'df_minmax_scaled' is ready from TASK-7.

# --- Data Selection ---

# Select two highly correlated features: Glucose and BMI
X = df_minmax_scaled[['Glucose', 'BMI']].values
y = diabetes_data['Outcome'].values

# Split the 2D data for visualization (using K=7 as a sample value)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Train a KNN model (using K=7)
k_value = 7
knn = KNeighborsClassifier(n_neighbors=k_value)
knn.fit(X_train, y_train)

# --- Meshgrid and Prediction for Boundary ---

# Define plot limits
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

# Create meshgrid of points across the plot space
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class for every point in the meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# --- Plotting the Diagram ---

# Define color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA']) # Background colors (Non-Diabetic, Diabetic)
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])  # Point colors

plt.figure(figsize=(12, 8))

# Plot the colored region (decision boundaries)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot the training data points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap_bold,
            edgecolor='k', s=40, label='Training Data')

# Plot the test data points
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cmap_bold,
            marker='x', s=100, linewidth=2, label='Test Data (X)')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title(f"KNN (K={k_value}) Voronoi Diagram on MinMax Scaled Data (Glucose vs BMI)")
plt.xlabel("Glucose (MinMax Scaled)")
plt.ylabel("BMI (MinMax Scaled)")
plt.legend()
plt.show()
```

```python
## TASK-9: Use K-Fold cross validation on all the above classification experiments and present an analysis of the results you obtain.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# --- 1. Data Preparation ---

# Assuming 'diabetes_data' is loaded from 'preprocessed_diabetes_data.csv'
# diabetes_data = pd.read_csv('preprocessed_diabetes_data.csv')

X_unscaled = diabetes_data.drop('Outcome', axis=1)
y = diabetes_data['Outcome']

# Standard Scaled Data
scaler_std = StandardScaler()
X_scaled_std = scaler_std.fit_transform(X_unscaled)

# MinMax Scaled Data
scaler_mm = MinMaxScaler()
X_scaled_mm = scaler_mm.fit_transform(X_unscaled)


# --- 2. K-Fold Cross-Validation Execution ---

cv_scores_unscaled = []
cv_scores_std = []
cv_scores_mm = []
k_range = range(1, 40)
cv_folds = 10 # Commonly used number of folds for CV

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)

    # Cross-validation for unscaled data
    scores_unscaled = cross_val_score(knn, X_unscaled, y, cv=cv_folds, scoring='accuracy')
    cv_scores_unscaled.append(scores_unscaled.mean())

    # Cross-validation for Standard Scaled data
    scores_std = cross_val_score(knn, X_scaled_std, y, cv=cv_folds, scoring='accuracy')
    cv_scores_std.append(scores_std.mean())

    # Cross-validation for MinMax Scaled data
    scores_mm = cross_val_score(knn, X_scaled_mm, y, cv=cv_folds, scoring='accuracy')
    cv_scores_mm.append(scores_mm.mean())


# --- 3. Analysis and Visualization ---

print("--- Cross-Validation Analysis ---")
print(f"Number of folds (CV): {cv_folds}\n")

# Unscaled Analysis
max_unscaled_score = max(cv_scores_unscaled)
k_unscaled_best = [k_range[i] for i, v in enumerate(cv_scores_unscaled) if v == max_unscaled_score]
print(f"Unscaled Data: Max CV Accuracy: {max_unscaled_score:.4f} at K={k_unscaled_best}")

# Standard Scaled Analysis
max_std_score = max(cv_scores_std)
k_std_best = [k_range[i] for i, v in enumerate(cv_scores_std) if v == max_std_score]
print(f"Standard Scaled Data: Max CV Accuracy: {max_std_score:.4f} at K={k_std_best}")

# MinMax Scaled Analysis
max_mm_score = max(cv_scores_mm)
k_mm_best = [k_range[i] for i, v in enumerate(cv_scores_mm) if v == max_mm_score]
print(f"MinMax Scaled Data: Max CV Accuracy: {max_mm_score:.4f} at K={k_mm_best}")


# Plotting the CV scores for comparison
plt.figure(figsize=(12, 8))
plt.plot(k_range, cv_scores_unscaled, marker='o', linestyle='--', label='Unscaled Data')
plt.plot(k_range, cv_scores_std, marker='o', linestyle='-', label='Standard Scaled Data')
plt.plot(k_range, cv_scores_mm, marker='o', linestyle=':', label='MinMax Scaled Data')

plt.title(f'10-Fold Cross-Validation Mean Accuracy vs. K Value (KNN)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel(f'Mean CV Accuracy (K={cv_folds} Folds)')
plt.legend()
plt.grid(True)
plt.savefig('knn_cv_analysis.png')
plt.close()

print("\nThe comparison plot of CV Mean Accuracy is saved as 'knn_cv_analysis.png'.")
```
